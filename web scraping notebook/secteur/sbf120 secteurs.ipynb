# Importation et fusion de la base

```python
sbf_120_base = pd.read_csv("sbf120 old.csv")
sbf_120_base
```

```python
sbf_120_base = sbf_120_base.rename(columns={"book_to_market": "book_value_per_share","sector": "industry", "indices": "index"})
sbf_120_base.head(3)
```

**Fusion des deux tables**

```python
sbf120 = pd.merge(sbf_120_base, secteur, on="firme")
sbf120
```

```python
sbf120.describe(include=object)
```

```python
sector = sbf120["sector"].unique()
sector
```

```python
# Correction des nom anglais de quelques secteurs 
sbf120["sector"].replace("Healthcare","Santé", inplace=True)
sbf120["sector"].replace("Financial","Finance", inplace=True)
sbf120["sector"].replace("Energy","Energie", inplace=True)
```

```python
sbf120["industry"].replace("Healthcare Facilities","Etablissements de santé", inplace=True)
sbf120["industry"].replace("Oil Well Services & Equipment","Puits de pétrole - services et équipement", inplace=True)
sbf120["industry"].replace("Misc. Financial Services","Services financiers (divers)", inplace=True)
```

```python
industry = sbf120["industry"].unique()
industry
```

```python
sbf120_backup = sbf120.copy()
sbf120.shape
```

```python
# Remplacer les millions dans la colonne des capitalisation boursière et du chiffre d'affaires
sbf120["market_cap"].replace("360.25M","0.32025B", inplace=True)
sbf120["market_cap"].replace("182.29M","0.18229B", inplace=True)
sbf120["market_cap"].replace("727.67M","0.72767B", inplace=True)
sbf120["market_cap"].replace("695.09M","0.69509B", inplace=True)
sbf120["market_cap"].replace("488.82M","0.48882B", inplace=True)
sbf120["market_cap"].replace("149.84M","0.14984B", inplace=True)
sbf120["market_cap"].replace("956.14M","0.95614B", inplace=True)
sbf120["market_cap"].replace("581.06M","0.58106B", inplace=True)
sbf120["market_cap"].replace("443.72M","0.44372B", inplace=True)
sbf120["market_cap"].replace("620.76M","0.62076B", inplace=True)
sbf120["market_cap"].replace("229.65M","0.22965B", inplace=True)
sbf120["revenue"].replace("756.39M","0.75639B", inplace=True)
sbf120["revenue"].replace("20.7M","0.0207B", inplace=True)
sbf120["revenue"].replace("27.31M","0.02731B", inplace=True)
sbf120["revenue"].replace("40.96M","0.04096B", inplace=True)
sbf120["revenue"].replace("492M","0.492B", inplace=True)
sbf120["revenue"].replace("283.86M","0.28386B", inplace=True)
sbf120["revenue"].replace("253.2M","0.2532B", inplace=True)
sbf120["revenue"].replace("554.27M","0.55427B", inplace=True)
sbf120["revenue"].replace("597.55M","0.59755B", inplace=True)
sbf120["revenue"].replace("369.64M","0.36964B", inplace=True)
```

```python
sbf120["market_cap"] = sbf120["market_cap"].str.rstrip("B").astype("float")/1.0
sbf120["revenue"] = sbf120["revenue"].str.rstrip("B").astype("float")/1.0
sbf120
```
